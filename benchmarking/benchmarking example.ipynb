## Methods

We use stochastic dominance approach to identify Monday effects.

If Monday has the lowest returns then it should be stochastically dominated by all other days.

We find first order and second order stochastic dominance here.

The first order test statistic is:

$ D1 = max_{k \neq M} sup_{x} \sqrt{N} [ D(x,k) - D(x,M)], $

$ D(x,k) = \frac{1}{N} \sum_{i=1}^{N} 1(X_{ki} \leq x) ,  where, k = M,T,W,Th,F $

The second order  test statistic is:

$ D2 = max_{k \neq M} sup_{x} \sqrt{N} [ D(x,k) - D(x,M)], $

$ D(x,k) = \frac{1}{N} \sum_{i=1}^{N} 1(X_{ki} \leq x)(x - X_{ki})  ,  where, k = M,T,W,Th,F $




### First Order Test Stat

```julia
grid =   [-26:0.07:45]
tic()
cvd = zeros(865,1)
gvd = zeros(1015,5)

for k in 1:5
    for j in 1:1015
        for i in 1:865
            if dta[i,k] <= grid[j,1]                
                cvd[i,1] = 1
            else cvd[i,1] = 0                    
            end
        end
        gvd[j,k] = sum(cvd)/865
    end
end

dts = zeros(4,1)

for i in 1:4
    dts[i,1] = maximum(gvd[:,i+1] - gvd[:, 1])
end

D1 = round(maximum(dts)* sqrt(865),3)
println("First Order Test Statistics is:", D1)

toc()
```

###### p value calculation:  We generate subsamples of size 500 from the full sample and calculate first order test stat, D1.  Then find the fraction of subsample test stats greater than full sample test stat.

### p value for first order test stat

```julia
tic()
ssize = 500

cvdp = zeros(865,1)
gvdp = zeros(1015,5)
dtsp = zeros(4,1)
dp1  = zeros(865 - ssize + 1, 1)

for p in 1 : 865 - ssize + 1
    dtap = dta[p:ssize - 1 + p,:]    
    for k in 1:5
        for j in 1:1015
            for i in 1:ssize
                if dtap[i,k] <= grid[j,1]
                    cvdp[i,1] = 1
                    else cvdp[i,1] = 0
                end
            end
            gvdp[j,k] = sum(cvdp)/ssize
        end
    end
    
    for n in 1:4
        dtsp[n,1] = maximum(gvdp[:, n+1] - gvdp[:, 1])
    end
    dp1[p:1] = round(maximum(dtsp)* sqrt(ssize),3)
end

ps1 = zeros(366,1)
for i in 1:366
    if dp1[i,1] > D1
        ps1[i,1] = 1
    else ps1[i,1] = 0
    end
end
pvalue1 = sum(ps1)/366

toc() 

println("p value for first order test stat is:", pvalue1)
```

Hence, we find the first order test stat to be significant. That is, we REJECT the null hypothesis of Monday Effects.

## Second Order Test Stat

```julia
tic()
cvd2 = zeros(865,1)
gvd2 = zeros(1015,5)

for k in 1:5
    for j in 1:1015
        for i in 1:865
            if dta[i,k] <= grid[j,1]                
                cvd2[i,1] = grid[j,1] - dta[i,k]
            else cvd2[i,1] = 0                    
            end
        end
        gvd2[j,k] = sum(cvd2)/865
    end
end


dts2 = zeros(4,1)

for i in 1:4
    dts2[i,1] = maximum(gvd2[:,i+1] - gvd2[:, 1])
end

D2 = round(maximum(dts2)* sqrt(865),3)

println("Second Order Test Statistics is:", D2)
toc()
```

### p value for second order test stat

```julia
tic()
ssize = 500

cvdp1 = zeros(865,1)
gvdp1 = zeros(1015,5)
dtsp1 = zeros(4,1)
dp2  = zeros(865 - ssize + 1, 1)

for p in 1 : 865 - ssize + 1
    dtap1 = dta[p:ssize - 1 + p,:]    
    for k in 1:5
        for j in 1:1015
            for i in 1:ssize
                if dtap1[i,k] <= grid[j,1]
                    cvdp1[i,1] = grid[j,1] - dtap1[i,k] 
                    else cvdp1[i,1] = 0
                end
            end
            gvdp1[j,k] = sum(cvdp1)/ssize
        end
    end
    
    for n in 1:4
        dtsp1[n,1] = maximum(gvdp1[:, n+1] - gvdp1[:, 1])
    end
    dp2[p:1] = round(maximum(dtsp1)* sqrt(ssize),3)
end

ps2 = zeros(366,1)
for i in 1:366
    if dp2[i,1] > D2
        ps2[i,1] = 1
    else ps2[i,1] = 0
    end
end
pvalue2 = sum(ps2)/366
toc()
println("p value for second order test stat is:", pvalue2)
```

Hence, we find the second order test stat to be significant. That is, we REJECT the null hypothesis of Monday Effects.
